# Calculating the Efficient Frontier of a Given ETF Portfolio

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tickers = ['VWO', 'VEA', '1328.T', 'XESC.MI', 'MEUD.PA', 'SCHV', 'ITOT', 'SPY']

etf = pd.DataFrame()

for t in tickers:
    etf[t] = yf.download(t)['Adj Close']
```

```python
etf.head()
```

Once the data is loaded and ready to go, we can first take a look at the join behavior of each one of the assets if they were normalized to 100. 

```python
normalized = (etf / etf.iloc[0]) * 100
normalized.plot(figsize=(14, 6))
```

## Calculating the Weights, Returns, Variance, and Volatility of the Portfolio

```python
etf_returns = np.log(etf/etf.shift())
etf_returns
```

```python
def get_weights(num_assets):
    num_assets = len(etf.columns)
    weights = np.random.random(num_assets)
    weights /= sum(weights)
    return weights
```

```python
weights = get_weights(etf)
```

```python
etf_cov = etf_returns.cov()
etf_cov
```

```python
etf_pfolio_returns = np.sum(weights * etf_returns.mean())*250
etf_pfolio_variance = np.dot(weights.T, np.dot(weights, etf_cov*250))
etf_pfolio_volatility = np.sqrt(etf_pfolio_variance)
```

## Simulating all Possible Combinations to Obtain the Efficient Frontier and Optimal Weights given Sharpe Ratio

```python
sim_weights = []
sim_returns= []
sim_vol = []

for sim in range(9000):
    weights = get_weights(etf)
    sim_weights.append(weights)
    sim_returns.append(np.sum(weights * etf_returns.mean())*250)
    sim_vol.append(np.sqrt(np.dot(weights.T, np.dot(weights, etf_cov*250))))
```

```python
sim_returns = np.array(sim_returns) 
sim_vol = np.array(sim_vol)
```

```python
etf_pfolio_returns_volatility = pd.DataFrame({'Returns': sim_returns, 'Volatility': sim_vol, 'Sharpe Ratio': sr})
sr= etf_pfolio_returns_volatility['Returns'] / etf_pfolio_returns_volatility['Volatility']
sr.max()
```

```python
etf_pfolio_returns_volatility
```

```python
plt.figure(figsize=(14,6))
plt.scatter(etf_pfolio_returns_volatility['Volatility'], etf_pfolio_returns_volatility['Returns'], c = sr)
plt.xlabel('Portfolio Expected Volatility')
plt.ylabel('Portfolio Expected Returns')
plt.show()
```

These whould be weights that optimize the Sharpe Ratio for the weights used in the simulation. 

```python
for i in sim_weights[np.argmax(sr)]:
    print(100*i)
```

But what would be the portfolios with the minimal variance and the maximum Sharpe Ratio?

In order to create these two portfolios, we are going to locate what value is the lowest for the volatility of the set of ETFs and which one stands atop for the Sharpe Ratio.

```python
# get the max and min values for Sharpe Ratio and volatility, respectively
min_volatility = etf_pfolio_returns_volatility['Volatility'].min()
max_sharpe = etf_pfolio_returns_volatility['Sharpe Ratio'].max()
```

```python
# get all the values that satisfy the conditions set 
sharpe_portfolio = etf_pfolio_returns_volatility.loc[etf_pfolio_returns_volatility['Sharpe Ratio'] == max_sharpe] 
min_var_portfolio = etf_pfolio_returns_volatility.loc[etf_pfolio_returns_volatility['Volatility'] == min_volatility]
```

```python
print(sharpe_portfolio)
print(min_var_portfolio)
```

According to the calculation, a Sharpe-maximizing portoflio would have an expected return of roughly 8% with a volatility of about 12%.
On the other hand, a portfolio seeking minimum variance would produce 7% in returns with something around 11% in volatility.
